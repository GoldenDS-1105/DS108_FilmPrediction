```python
# Đổi tên cột ở imdb
df_imdb = df_imdb.rename(columns={
    'vote_avg': 'vote_average',
    'vote_cnt': 'vote_count',
    'lang': 'original_language',
})
```

```python
# Xóa những cột không giống
df_imdb.drop(columns=['metascore', 'certificate', 'pop'], inplace=True)

df_tmdb.drop(columns=['top_cast','production_companies','popularity','keywords'], inplace=True)
```

```python
# Chia cột vote_average imdb cho 1000
df_imdb['vote_average'] = df_imdb['vote_average'] / 1000
df_imdb
```

```python
# Chuyển runtime về kiểu int, vote_average về float
# runtime → int
df_imdb['runtime'] = pd.to_numeric(df_imdb['runtime'], errors='coerce').astype('Int64')

# vote_average → float (giữ nguyên)
df_imdb['vote_average'] = pd.to_numeric(df_imdb['vote_average'], errors='coerce')
```

```python
df_imdb.info()
```

```python
df_imdb
```

```python
# Gộp 2 file lại
df_all = pd.concat([df_imdb, df_tmdb], ignore_index=True)
```

```python
# Xóa duplicate theo title và year
df_all['title'] = df_all['title'].str.lower().str.strip()
df_all['year'] = pd.to_datetime(df_all['release_date'], errors='coerce').dt.year

df_all = df_all.drop_duplicates(subset=['title', 'year'])
```
